# Modern Operations Research with Python

**Optimization • Machine Learning • Data-Driven Decision Making**

A practical eBook combining classical Operations Research techniques with modern Machine Learning approaches using Python and Quarto.

## Current Book Structure

- **Chapter 1**: Introduction  
- **Chapter 2**: Linear Programming  
- **Chapter 3**: Network Optimization  
- **Chapter 4**: Integer Programming  
- **Chapter 5**: Nonlinear Optimization  
- **Chapter 6**: Stochastic Programming ← **Current focus**  
- **Chapter 7**: Hybrid OR-ML Approaches  
- **Chapter 8+**: Reinforcement Learning, Case Studies, etc.

## Technologies Used

- **Quarto** — Book publishing framework
- **Python 3.12** (via `or-ml-ebook-env`)
- **PuLP + GLPK** — Optimization
- **SciPy** — Nonlinear optimization
- **scikit-learn** — Machine Learning
- **Jupyter Notebooks** — Interactive development

## Quick Start

```bash
cd /Users/troyaltus/data_projects/or-ml-ebook
or-ml-ebook-env
quarto render --to html
quarto preview
```

## Development Workflow

1. Experiment in `notebooks/` using `.ipynb` files
2. Convert to clean chapters using:
   ```bash
   quarto convert notebooks/XX-name.ipynb --output book/XX-name.qmd
   ```
3. Polish the `.qmd` file
4. Update `_quarto.yml`
5. Test rendering and push to GitHub
