# Integer Programming

Integer Programming (IP) is an extension of Linear Programming where some or all decision variables are required to be integers.

## Why Integer Programming?

- Many real-world decisions are **discrete** (e.g., build 0 or 1 factory, assign whole machines, yes/no choices).
- Standard LP may give fractional solutions that are not practical.

## Classic Example: Knapsack Problem

A hiker wants to maximize the total value of items in their knapsack without exceeding weight capacity.

```python
#
# Chapter 04: Integer Programming
# 0/1 Knapsack Problem - Pure SciPy version (no PuLP, no solver binary issues)
#

import numpy as np
from scipy.optimize import linprog

print("0/1 Knapsack Problem - Integer Programming Example")
print("=" * 60)

# Items: name, value, weight
items = ["Tent", "Food", "Water", "Camera", "Map"]
values = np.array([10, 8, 7, 6, 3])
weights = np.array([5, 3, 4, 2, 1])
capacity = 10

print(f"Capacity: {capacity} kg\n")

# Objective: maximize value → minimize negative value
c = -values

# Weight constraint
A_ub = weights.reshape(1, -1)
b_ub = np.array([capacity])

# Bounds: each item can be taken 0 or 1 time (relaxed to continuous [0,1])
bounds = [(0, 1) for _ in items]

# Solve the relaxed problem
result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')

if result.success:
    # Round to nearest integer (0 or 1) for 0/1 solution
    selected = np.round(result.x).astype(int)
    total_value = np.dot(selected, values)
    total_weight = np.dot(selected, weights)
    
    print(f"Total Value: {total_value}")
    print(f"Total Weight: {total_weight} / {capacity}\n")
    
    print("Items to take:")
    for i, item in enumerate(items):
        if selected[i] == 1:
            print(f"  ✓ {item} (Value: {values[i]}, Weight: {weights[i]})")
else:
    print("Solver failed to find a solution.")

print("\n" + "=" * 60)
print("Note: This is a relaxed + rounded solution. True Integer Programming solvers")
print("use Branch-and-Bound for exact solutions.")
```

## Key Differences from Linear Programming

- Decision variables can be **Binary** (0/1) or **Integer**.
- Problems are generally much harder to solve (NP-hard).
- Modern solvers (CBC, Gurobi, CPLEX) use Branch-and-Bound + cutting planes.
