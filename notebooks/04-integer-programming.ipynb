# Chapter 04: Integer Programming

Integer Programming (IP) extends Linear Programming by requiring some or all decision variables to be **integers** (often binary 0 or 1).

## Why Integer Programming?

Many real-world decisions are discrete:
- Build a factory or not
- Assign whole machines or people
- Choose yes/no options

Standard Linear Programming may give fractional solutions that are not practical.

## Classic Example: 0/1 Knapsack Problem

A hiker wants to maximize the total value of items carried in a knapsack without exceeding the weight limit.

```python
import numpy as np
from scipy.optimize import linprog

print("0/1 Knapsack Problem Solution")
print("=" * 50)

# Items: name, value, weight
items = ["Tent", "Food", "Water", "Camera", "Map"]
values = np.array([10, 8, 7, 6, 3])
weights = np.array([5, 3, 4, 2, 1])
capacity = 10

# Objective: maximize value → minimize negative value
c = -values

# Weight constraint
A_ub = weights.reshape(1, -1)
b_ub = np.array([capacity])

# Bounds: each item is 0 or 1 (relaxed to continuous for simplicity)
bounds = [(0, 1) for _ in items]

# Solve
result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')

if result.success:
    selected = np.round(result.x).astype(int)
    total_value = np.dot(selected, values)
    total_weight = np.dot(selected, weights)
    
    print(f"Total Value: {total_value}")
    print(f"Total Weight: {total_weight} / {capacity}\n")
    
    print("Items to take:")
    for i, item in enumerate(items):
        if selected[i] == 1:
            print(f"  ✓ {item} (Value: {values[i]}, Weight: {weights[i]})")
else:
    print("Solver failed")
```

## Key Concepts

- Decision variables are **binary** (0 or 1) or integer.
- Problems are generally **NP-hard** — harder than pure Linear Programming.
- Modern solvers use Branch-and-Bound combined with cutting planes.
- In practice, PuLP + CBC or Gurobi are commonly used for exact solutions.
