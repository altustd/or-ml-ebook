# Nonlinear Optimization

**Chapter 6**

While linear and integer programming assume linear relationships, many real-world operations research problems involve nonlinear objectives or constraints. Examples include quadratic costs, diminishing returns, geometric relationships, or complex physical systems.

This chapter introduces **nonlinear programming (NLP)**, discusses the challenges of non-convexity, and demonstrates practical solution techniques using SciPy and other tools.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize, differential_evolution
import seaborn as sns

sns.set(style="whitegrid")
np.random.seed(42)

print("Nonlinear optimization libraries loaded")
```

## 6.1 Why Nonlinear Optimization?

Nonlinear problems arise naturally in OR when:
- Costs are quadratic (e.g., inventory holding costs)
- Revenue exhibits diminishing returns
- Constraints involve products, ratios, or geometric terms
- We optimize machine learning model hyperparameters

Unlike linear programs, nonlinear problems can have **multiple local optima**, making global optimization significantly harder.

## 6.2 Unconstrained Nonlinear Optimization – Rosenbrock Function

The Rosenbrock function is a classic test problem for nonlinear optimizers. It has a long, narrow valley that makes convergence difficult for many algorithms.

```python
def rosenbrock(x):
    return (1 - x[0])**2 + 100 * (x[1] - x[0]**2)**2

# Initial guess
x0 = np.array([-1.2, 1.0])

result = minimize(rosenbrock, x0, method='BFGS')

print("=== Rosenbrock Function Optimization ===")
print(f"Optimal solution: x = {result.x}")
print(f"Minimum value   : {result.fun:.6f}")
print(f"Success         : {result.success}")
```

## 6.3 Constrained Nonlinear Optimization

Most practical problems have constraints. Here we minimize a nonlinear objective subject to nonlinear inequality constraints.

```python
def objective(x):
    return x[0]**2 + x[1]**2 + x[0]*x[1]

def constraint1(x):
    return x[0] + x[1] - 5   # x0 + x1 >= 5

cons = {'type': 'ineq', 'fun': constraint1}

bounds = [(-10, 10), (-10, 10)]
x0 = np.array([1.0, 1.0])

result = minimize(objective, x0, bounds=bounds, constraints=cons, method='SLSQP')

print("=== Constrained Nonlinear Optimization ===")
print(f"Optimal solution: x = [{result.x[0]:.4f}, {result.x[1]:.4f}]")
print(f"Objective value : {result.fun:.4f}")
print(f"Constraint value: {constraint1(result.x):.4f} (should be ≥ 0)")
```

## 6.4 Global Optimization with Differential Evolution

Gradient-based methods like SLSQP or BFGS can get stuck in local minima. **Differential Evolution** is a powerful population-based global optimizer that does not require gradients.

```python
def multimodal_objective(x):
    return np.sin(3*x[0]) * np.cos(5*x[1]) + x[0]**2 + x[1]**2

bounds = [(-2, 2), (-2, 2)]

result_de = differential_evolution(multimodal_objective, bounds, workers=1)

print("=== Global Optimization with Differential Evolution ===")
print(f"Best solution found: x = [{result_de.x[0]:.4f}, {result_de.x[1]:.4f}]")
print(f"Best objective value : {result_de.fun:.6f}")
```

## 6.5 Key Takeaways

- Nonlinear problems can have **multiple local optima**.
- Gradient-based methods (BFGS, SLSQP) are fast but sensitive to starting points.
- Global optimizers (Differential Evolution, Bayesian optimization) are more robust but slower.
- Choosing the right solver and providing good initial guesses is critical in practice.
