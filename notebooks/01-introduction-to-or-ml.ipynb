## Simple Linear Programming Example (Classic OR)

```python
import pulp

# Production planning problem
prob = pulp.LpProblem("Production_Planning", pulp.LpMaximize)

# Variables
x1 = pulp.LpVariable("Product_A", lowBound=0, cat='Integer')
x2 = pulp.LpVariable("Product_B", lowBound=0, cat='Integer')

# Objective: Maximize profit
prob += 25 * x1 + 30 * x2, "Total_Profit"

# Constraints
prob += 3 * x1 + 4 * x2 <= 240, "Machine_Hours"
prob += 2 * x1 + 1 * x2 <= 100, "Labor_Hours"

prob.solve(pulp.PULP_CBC_CMD(msg=False))

print(f"Optimal Production:")
print(f"Product A: {pulp.value(x1):.0f} units")
print(f"Product B: {pulp.value(x2):.0f} units")
print(f"Maximum Profit: ${pulp.value(prob.objective):.2f}")
```

## Machine Learning Component: Demand Prediction

```python
# Simple synthetic demand prediction example
weeks = np.arange(1, 53)
demand = 120 + 15 * np.sin(weeks * np.pi / 12) + np.random.normal(0, 12, 52)

plt.figure(figsize=(10, 5))
plt.plot(weeks, demand, 'o-', label='Weekly Demand')
plt.axhline(y=120, color='red', linestyle='--', label='Average Demand')
plt.title('Stochastic Demand Pattern')
plt.xlabel('Week')
plt.ylabel('Demand (units)')
plt.legend()
plt.show()

print(f"Average demand: {demand.mean():.1f} units")
print(f"Demand variability (std): {demand.std():.1f} units")
```
