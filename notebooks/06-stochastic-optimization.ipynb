# Stochastic Optimization

Stochastic Optimization deals with problems where uncertainty or randomness plays a key role.

## Why Stochastic Optimization?

In real life, many parameters are not known with certainty:
- Demand fluctuates
- Prices change randomly
- Machine failures occur unpredictably
- Travel times vary

Stochastic models help make better decisions under uncertainty.

## Classic Example: The Newsvendor Problem

A newsstand owner must decide how many newspapers to order each day. Demand is random. Ordering too many wastes money; ordering too few loses potential sales.

```python
import numpy as np

# Parameters
cost = 2.0           # cost per newspaper
selling_price = 5.0  # selling price
salvage = 0.5        # salvage value for unsold papers

# Demand scenarios and probabilities
demand_scenarios = np.array([50, 60, 70, 80, 90, 100])
probabilities = np.array([0.1, 0.15, 0.2, 0.25, 0.2, 0.1])

def expected_profit(order_quantity):
    profit = 0
    for d, p in zip(demand_scenarios, probabilities):
        sales = min(order_quantity, d)
        unsold = max(0, order_quantity - d)
        revenue = sales * selling_price
        salvage_value = unsold * salvage
        total_cost = order_quantity * cost
        profit += p * (revenue + salvage_value - total_cost)
    return profit

# Test different order quantities
quantities = range(40, 121, 5)
profits = [expected_profit(q) for q in quantities]

best_idx = np.argmax(profits)
best_quantity = quantities[best_idx]
best_profit = profits[best_idx]

print(f"Optimal order quantity: {best_quantity} newspapers")
print(f"Expected daily profit: ${best_profit:.2f}")

# Show profit for a few quantities
print("\nProfit for selected quantities:")
for q in [60, 70, 80, 90]:
    print(f"Order {q}: Expected profit = ${expected_profit(q):.2f}")
```

## Key Takeaways

- Stochastic optimization helps make robust decisions under uncertainty.
- The Newsvendor problem is a fundamental building block for inventory and supply chain models.
- Common approaches include scenario-based methods, chance-constrained programming, and robust optimization.
