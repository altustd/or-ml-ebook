# Linear Programming

Foundations of optimization: Modeling and solving linear programs with PuLP in Python.

```python
import pulp

print("PuLP version:", pulp.__version__)
```

## Production Planning Problem (2 Products)

```python
# Create the linear programming problem
prob = pulp.LpProblem("Production_Planning_Two_Products", pulp.LpMaximize)

# Decision variables
x1 = pulp.LpVariable("Product_A", lowBound=0, cat="Integer")
x2 = pulp.LpVariable("Product_B", lowBound=0, cat="Integer")

# Objective function: Maximize profit
prob += 25 * x1 + 30 * x2, "Total_Profit"

# Constraints
prob += 3 * x1 + 4 * x2 <= 240, "Machine_Hours"
prob += 2 * x1 + 1 * x2 <= 100, "Labor_Hours"

# Demand limits
prob += x1 <= 60, "Max_Demand_A"
prob += x2 <= 50, "Max_Demand_B"

# Solve the model
status = prob.solve(pulp.PULP_CBC_CMD(msg=False))

print("Optimal Production Plan:")
print(f"Product A : {pulp.value(x1):.0f} units")
print(f"Product B : {pulp.value(x2):.0f} units")
print(f"Total Profit : ${pulp.value(prob.objective):.2f}")
```

## Resource Utilization

```python
print("\nResource Usage:")
print(f"Machine Hours : {3*pulp.value(x1) + 4*pulp.value(x2):.0f} / 240")
print(f"Labor Hours   : {2*pulp.value(x1) + 1*pulp.value(x2):.0f} / 100")
```

## Shadow Prices

```python
print("\nShadow Prices (Marginal Value of Resources):")
for name, c in prob.constraints.items():
    print(f"{name:20s}: {c.pi:.4f}")
```
