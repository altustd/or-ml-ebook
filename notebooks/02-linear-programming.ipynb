# Chapter 02: Linear Programming

Linear Programming (LP) is one of the most important techniques in Operations Research.

## Classic Example: Transportation Problem

A company has two factories (F1 and F2) and three warehouses (W1, W2, W3). 
Goal: Minimize total shipping cost while meeting supply and demand.

```python
import numpy as np
from scipy.optimize import linprog

print("Transportation Problem Solution")
print("=" * 60)

# Cost coefficients: F1W1, F1W2, F1W3, F2W1, F2W2, F2W3
c = np.array([5, 8, 4, 7, 6, 9])

# Inequality constraints (A_ub x <= b_ub)
A_ub = np.array([
    [1, 1, 1, 0, 0, 0],   # F1 supply <= 100
    [0, 0, 0, 1, 1, 1]    # F2 supply <= 150
])
b_ub = np.array([100, 150])

# Demand constraints (converted to <= form)
demand_A = np.array([
    [-1, 0, 0, -1, 0, 0],   # W1 demand >= 80
    [ 0,-1, 0,  0,-1, 0],   # W2 demand >= 120
    [ 0, 0,-1,  0, 0,-1]    # W3 demand >= 50
])

A_ub = np.vstack([A_ub, demand_A])
b_ub = np.append(b_ub, [-80, -120, -50])

# Solve the linear program
result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=(0, None), method='highs')

if result.success:
    print(f"Total Transportation Cost: ${result.fun:.2f}\n")
    
    # Reshape solution into 2 factories x 3 warehouses
    shipments = result.x.reshape(2, 3)
    
    print("Optimal Shipping Plan:")
    print("From \\ To   W1   W2   W3")
    print("-" * 35)
    for i, f in enumerate(["F1", "F2"]):
        row = [f"{shipments[i,j]:.0f}" for j in range(3)]
        print(f"{f:<8} {row[0]:>3}  {row[1]:>3}  {row[2]:>3}")
else:
    print("Solver failed to find an optimal solution.")
```

## Key Takeaways

- Linear Programming is highly effective for allocation and logistics problems.
- `scipy.optimize.linprog` with the `highs` method is reliable and easy to use.
- Always check `result.success` to ensure the solver found a valid solution.
