# Chapter 07: Reinforcement Learning in Operations Research

Reinforcement Learning (RL) is a powerful machine learning paradigm where an agent learns to make sequential decisions by interacting with an environment to maximize cumulative reward.

## Why Combine RL with Operations Research?

- OR excels at optimization with known models
- RL excels at learning optimal policies in complex, uncertain, or dynamic environments
- Together they can solve problems that are too complex for traditional OR alone

## Example: Dynamic Inventory Management with RL

A retailer must decide how much to order each period under uncertain demand and holding/backorder costs.

```python
import numpy as np
import random

class InventoryEnv:
    def __init__(self, max_inventory=20, demand_mean=8, holding_cost=1, backorder_cost=5):
        self.max_inventory = max_inventory
        self.demand_mean = demand_mean
        self.holding_cost = holding_cost
        self.backorder_cost = backorder_cost
        self.reset()
    
    def reset(self):
        self.inventory = random.randint(5, 15)
        self.total_reward = 0
        return self.inventory
    
    def step(self, order_quantity):
        demand = max(0, int(np.random.normal(self.demand_mean, 3)))
        
        # Update inventory
        self.inventory = min(self.max_inventory, self.inventory + order_quantity)
        
        # Calculate sales and backorders
        sales = min(self.inventory, demand)
        backorder = max(0, demand - self.inventory)
        self.inventory = max(0, self.inventory - demand)
        
        # Calculate reward (negative cost)
        holding_cost = self.inventory * self.holding_cost
        backorder_cost = backorder * self.backorder_cost
        reward = - (holding_cost + backorder_cost)
        
        self.total_reward += reward
        
        done = False  # single period for simplicity
        return self.inventory, reward, done

# Simple Q-learning agent for illustration
env = InventoryEnv()
print("Environment created.")
print(f"Initial inventory: {env.inventory}")

# Simulate one episode with random policy
total_reward = 0
for step in range(30):
    action = random.randint(0, 10)  # order 0 to 10 units
    state, reward, done = env.step(action)
    total_reward += reward

print(f"Total reward after 30 periods: {total_reward:.1f}")
```

## Key Takeaways

- Reinforcement Learning is well-suited for sequential decision-making under uncertainty.
- It can complement traditional OR methods in dynamic environments.
- Common applications include inventory control, routing, scheduling, and pricing.

In more advanced setups, we use Deep Q-Networks (DQN), Policy Gradient methods, or Actor-Critic algorithms.
